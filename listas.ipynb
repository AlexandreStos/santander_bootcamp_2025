### Listas aninhadas

Podemos armazenar listas dentro de listas, com isso podemos criar tabelas, acessando através de índices de linhas e colunas, ou seja, matriz


#### Matriz

```python
matriz = [
    [1, "a", 2], 
    ["b", 3, 4], 
    [6, 5, "c"],        
]


print(matriz [0]) # linha zero  (primeira) 
print(matriz [0] [0]) # linha 1 e elemento 1
print(matriz [0] [-1])# linha zero e ultimo elemento   
print(matriz [-1] [-1]) # último elemento da última linha   
```

### Fatimento

#### Podemos pegar também partes dos elementos, passando o índice inicial e final, e/ou o step
#### - Devemos usar os ":" para "acessar"
#### - Na maioria das vezes devemos passar pelo menos o índice inicial ou final
### ** Posição final não é retornada **

#### Ex.: 

```python
lista = ["p", "y", "t", "h", "o", "n"]

print(lista[2:]) 
print(lista[:2])
print(lista[1:3])
print(lista[0:3:2])
print(lista[::]) ### usado para copiar ### excessão 
print(lista[::-1]) ### copia ao contrário
```

### Compreensão de listas/filtros

#### - Usado para criar uma lista nova a partir de uma lista existente, ou modificada

#### Filtro versão 1

```python
numeros = [1,30,21, 2,9,65,34]
pares = []

for x in numeros:
    if x % 2 == 0:
        pares.append(x)

print(pares)
```

### Filtro versão 2. *list comprehension*


```python
pares = [x for x in numeros if x %2 == 0]
print(pares)
```

### Modificando os valores versão 1

```python
quadrado = []
```

```python
for x in numeros:
    quadrado.append(x**2)
print(quadrado)
```

### Modificando os valores versão 2

```python
quadrado = [x ** 2 for x in numeros]
print(quadrado)
```

### .copy
##### Diferente do R mesmo que eu copie com outro nome a modificação da lista acontece em ambos os objetos, por isso é necessário realizar o uso do método .copy

```python
lista.copy()
```

```python
lista = [1, 2, 3]
lista2 = lista  # Não é uma cópia, apenas uma nova referência
lista2.append(4)
print(lista)  # Saída: [1, 2, 3, 4] (a original foi alterada!)

# Usando copy()
lista = [1, 2, 3]
lista2 = lista.copy()  # Cria uma cópia independente
lista2.append(4)
print(lista)   # Saída: [1, 2, 3] (original não foi alterada)
print(lista2)  # Saída: [1, 2, 3, 4]
```

###.extend

O método .extend() adiciona cada elemento de um iterável (como outra lista) ao final da lista, enquanto o .append() adiciona o objeto inteiro como um único elemento. Não adiciona como uma lista aninhada.

```python
lista = [1, 2, 3]
lista.append([4, 5])
print(lista)  # Saída: [1, 2, 3, [4, 5]]
```

#### Exemplo com .extend()

```python
lista = [1, 2, 3]
lista.extend([4, 5])
print(lista)  # Saída: [1, 2, 3, 4, 5]
```

### Resumo
#### .append(): Adiciona um elemento (pode ser uma lista, mas ela entra como um único item).
#### .extend(): Adiciona vários elementos, um a um, de um iterável.

```python
### .pop
```

O método .pop() em Python segue o conceito de pilha, que pode ser entendido como uma pilha de pratos: você adiciona pratos no topo e retira sempre do topo, ou seja, o último a entrar é o primeiro a sair (LIFO - Last In, First Out).

Exemplo com pilha de pratos

```python
pratos = ['prato1', 'prato2', 'prato3']

# Adicionando pratos no topo (push)
pratos.append('prato4')
print(pratos)  # Saída: ['prato1', 'prato2', 'prato3', 'prato4']

# Retirando prato do topo (pop)
ultimo_prato = pratos.pop()
print(ultimo_prato)  # Saída: prato4
print(pratos)        # Saída: ['prato1', 'prato2', 'prato3']
```

#### Exemplo 2: escolhendo o elemento

```python
lista = ['maçã', 'banana', 'laranja', 'uva']
elemento = lista.pop(1)  # Remove o elemento no índice 1 (banana)
print(elemento)  # Saída: banana
print(lista)     # Saída: ['maçã', 'laranja', 'uva']
```

### .sort

O método .sort() em Python ordena os elementos de uma lista diretamente, modificando a própria lista. Por padrão, ele organiza os elementos em ordem crescente (ascendente), mas você pode alterar para ordem decrescente usando o parâmetro reverse=True

#### Exemplo básico

```python
numeros = [3, 1, 4, 1, 5]
numeros.sort()
print(numeros)  # Saída: [1, 1, 3, 4, 5]
```

#### Ordem decrescente (ordena e inverte)

```python
numeros = [3, 1, 4, 1, 5]
numeros.sort(reverse=True)
print(numeros)  # Saída: [5, 4, 3, 1, 1]
```

#### Ordenação personalizada

Você também pode usar o parâmetro key para definir critérios de ordenação, como ordenar strings pelo tamanho:

```python
palavras = ['kiwi', 'apple', 'banana']
palavras.sort(key=len)
print(palavras)  # Saída: ['kiwi', 'apple', 'banana'] (ordenado pelo comprimento)
```

### Combinando key com reverse

```python
palavras = ['kiwi', 'apple', 'banana']
palavras.sort(key=len, reverse=True)
print(palavras)  # Saída: ['kiwi', 'apple', 'banana'] (ordenado pelo comprimento)
```
