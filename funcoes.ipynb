### funções

#### A função é um bloco de código nomeado e padronizado com o objetivo de realizar alguma atividade no código geral, tornando mais legível e possibilitando o reaproveitamento do bloco de código, ou seja, programando de forma estruturada.

#### Usamos a palavra reservada def para definir o uso da função

```python
def exibir_mensagem():
    print("Olá mundo!")
```

```python
exibir_mensagem()
```

```python
def exibir_mensagem(nome):
    print(f"Olá {nome}")
```

```python
exibir_mensagem("Alexandre")
```

```python
def exibir_mensagem(nome="Anônimo"):  ### aqui declaramos automáticamente o nome como anônimo, evitando o erro, caso não digitado
    print(f"Olá {nome}")
```

```python
exibir_mensagem()
```

### Retornando valores
  #### Devemos usar a palavra reservada "return"

```python
def calcular_total(numeros): ## precisa ser uma lista 
    return sum(numeros)
```

```python
calcular_total([10,2])
```

```python
def ant_suce(numero):
    antecessor = numero - 1
    sucessor = numero + 1
    return print(f"O antecessor do número é {antecessor} e o sucessor é {sucessor }")
```

```python
ant_suce(10)
```

### Argumentos nomeados

### *Args e **kwargs (convenção)

Podemos combinar parâmetros obrigatórios com args e kwargs. Quando esses sào definidos (*args e **kwargs), o metódo recebe os valores como tuplas e dicionários

Args é tupla (*)
Kwargs é chave e valor(**)

```python
def exibir_poema(data_extenso, *args, **kwargs):
    texto = "\n".join(args)
    meta_dados = "\n".join([f"{chave.title()}:{valor}" for chave, valor in kwargs.items()])
    mensagem = f'{data_extenso}\n\n{texto}\n\n{meta_dados}'
    print(mensagem)
    
```

```python
exibir_poema("Sexta", "Bla bla", "lalalalala", autor = "Eu", ano = 2025)
```

### Objetos de primeira classe

O que são objetos de primeira classe em Python?
Em programação, um objeto de primeira classe (first-class object) é qualquer entidade que pode ser tratada da mesma forma que outras entidades comuns da linguagem, ou seja:
- Pode ser atribuída a uma variável
- Pode ser passada como argumento para uma função
- Pode ser retornada por uma função
- Pode ser armazenada em estruturas de dados (listas, dicionários, etc.)

#### Posso passar a função dentro de função como argumento ou passar dentro de objetos

```python
def soma(a, b):
    return a + b

# 1. Atribuir a uma variável
minha_funcao = soma
print(minha_funcao(3, 4))  # Saída: 7

# 2. Passar como argumento
def executar(funcao, x, y):
    return funcao(x, y)

print(executar(soma, 5, 6))  # Saída: 11

# 3. Retornar de uma função
def criar_multiplicador(fator):
    def multiplicar(x):
        return x * fator
    return multiplicar

dobro = criar_multiplicador(2)
print(dobro(10))  # Saída: 20

# 4. Armazenar em estrutura de dados
operacoes = [soma, criar_multiplicador(3)]
print(operacoes[0](2, 3))  # 5
print(operacoes[1](4))     # 12
```

### Escopo global (fora da função) ou local (dentro da função)

Em Python, o **escopo** determina onde uma variável é visível e pode ser acessada no código. Existem dois escopos principais:

### Tipos de Escopo

| Escopo      | Onde é criado                          | Visibilidade                                       | Exemplo de criação                          |
|-------------|----------------------------------------|----------------------------------------------------|---------------------------------------------|
| **Local**   | Dentro de uma função ou lambda         | Apenas dentro dessa função                         | Parâmetro ou atribuição dentro da função     |
| **Global**  | Fora de qualquer função (nível módulo) | Em todo o módulo e dentro de funções (para leitura)| Variável criada no corpo principal do script|

### Regra LEGB
Quando Python procura uma variável, ele segue esta ordem:

1. **L**ocal (dentro da função atual)  
2. **E**nclosing (funções externas – útil em closures)  
3. **G**lobal (nível do módulo)  
4. **B**uilt-in (nomes nativos do Python como `print`, `len`)

### Comportamento ao modificar variáveis

| Ação dentro da função                  | Sem `global`                          | Com `global`                          |
|----------------------------------------|---------------------------------------|---------------------------------------|
| Apenas **ler** a variável              | Funciona (pega o valor global)        | Funciona                              |
| **Atribuir** valor à variável          | Cria uma nova variável **local**      | Modifica a variável **global**        |

### Exemplo prático



```python
contador = 10  # global

def test():
    global contador      # avisa que vamos usar/modificar a global
    print(contador)      # agora lê a global → 10
    contador = 20        # modifica a global

test()
print(contador)          # Saída: 20
```
