### Operadores de identidade 


#### Utilizados para identificar se os objetos ocupam o mesmo espaço na memória, ou seja, são a mesma informação.


#### Exemplo:

```python
curso = "Curso de Python"
nome_curso = curso
saldo, limite = 200, 200
```

```python
curso is nome_curso
```

```python
curso is not nome_curso
```

```python
saldo is limite
```

Comparação com o operador ==

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True  (valores iguais)
print(a is b)  # False (objetos diferentes)
```

Python faz cache de alguns inteiros pequenos, então is pode dar True em casos específicos.
O que só reforça: use == para comparar valores e deixe is para checagens de identidade, como is None.
