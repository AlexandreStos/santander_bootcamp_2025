#### Operadores de Atribuição

```python
saldo = 50
print(saldo)
```

### Atribuição com adição

#### Maneira dreduzida, sem a necessidade de usar o nome da variável para atribuir

```python
saldo += 20
print(saldo)
```

### Atribuição com subtração

```python
saldo -= 20
print(saldo)
```

### Atribuição com multiplicação

```python
saldo *= 20
print(saldo)
```

### Atribuição com divisão

```python
saldo /= 20
print(saldo)
```

### Atribuição com divisão inteira 

```python
saldo //= 3.5
print(saldo)
```

### Atribuição com modulo (resto da divisão)

```python
saldo %= 3.5
print(saldo)
```

### Atribuição com exponenciação

```python
saldo **= 3.5
print(saldo)
```
