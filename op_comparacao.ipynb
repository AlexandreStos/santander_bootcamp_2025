### Operadores de comparação

#### - Utilizados para fazer comparações entre valores

```python
saldo = 100
saque = 20
```

##### Comparação de igualdade 

```python
print(saque == saldo)
```

##### Comparação de diferença

```python
print(saque != saldo)
```

##### Comparação se um valor é maior que outro

```python
print(saque > saldo)
```

##### Comparação se um valor é menor que outro

```python
print(saque < saldo)
```

##### Comparação se um valor é maior OU IGUAL que outro

```python
print(saque >= saldo)
```

```python
##### Comparação se um valor é menor OU IGUAL que outro
```

```python
print(saque <= saldo)
```
