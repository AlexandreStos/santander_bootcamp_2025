# Bancos de dados relacionais, SQL e ETL

#### Bancos de dados é uma coleção organizada e estruturada, armazenada de forma eletrônica (na maioria das vezes)

#### Tipos de bancos de dados:
#### - Relacionais ou SQL 
#### - Não relacionais/NoSQL
#### - Orientado a objetos
#### - Hierárquicos

### Sistemas de gerenciamento de banco de dados (SGBD)

#### Funcionalidades básicas

#### -  **C**reat
#### -  **R**read
#### -  **U**pdate
#### -  **D**elete

### Estrutura de banco de dados 

#### É formado por um conjunto de tabelas (ou mais de uma)

#### As tabelas são compostas por colunas, variáveis ou features, e as tuplas, linhas, observações ou registros 

Os relacionamentos entre as tabelas dependendem das chaves primárias e estrangeiras 

### Características de bancos de dados relacionais 

#### &star; Relacionamento entre tabelas 
#### &star; Linguagem de consultas estruturadas (SQL)
#### &star; Integridade referencial
#### &star; Normalização de dados
#### &star; Segurança 
#### &star; Flexibilidade e extensibilidade
#### &star; Suporte a transações ACID (atomicidade, consistência, isolamento, durabilidade)


## Conceitos básicos 

### SQL - Structured Query Language

##### Permite consultas, alterar dados ou filtrar

### DQL - Linguagem de Consulta de dados 
   - SELECT

### DML - Linguagem de manipulação de dados 
   - INSERT, UPDATE e DELETE

#### DDL - Linguagem de definição de dados 
   - CREAT, ALTER, DROP

#### DCL - Linguagem de Controle de Dados 
   - GRANT, REVOKE

#### DTL - Linguagem de transação de dados
  - BEGIN, COMMIT, ROLLBACK

### Sintaxe básica: Nomeclatura 

 - Os nomes devem começar com uma letra ou com um caractere de sublinhado (_)
 - Os nomes podem conter letras, números ou caracteres
 - Case sensitivity - Sensibilidade a maiúsculas e minúsculas (depende da linguagem e banco)

*Python é case sensitive. Isso significa que identificadores (nomes de variáveis, funções, classes, módulos) diferenciam maiúsculas de minúsculas.*
