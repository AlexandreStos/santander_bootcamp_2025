### Tuplas

 Tuplas são estruturas imutáveis, após criada nao pode ser mudada até o final do programa.
- Para criar podemos usar parenteses e inserir os objetos ou usar a função de base tuple()

```python
frutas = ("laranja", "pera", "uva",) ## virgula no final
letras = tuple("python")
numeros = tuple([1,2,3,4])
pais = ("Brasil",) ## virgula no final
```

#### *Podemos armazenar outros tipos de objetos nas tuplas, podendo armazenar outras tuplas aninhadas, formando tabelas.*

```python
matriz = (
    (1, "a", 2), 
    ("b", 3, 4), 
    (6, 5, "c"),        
)


print(matriz [0]) # linha zero  (primeira) 
print(matriz [0] [0]) # linha 1 e elemento 1
print(matriz [0] [-1])# linha zero e ultimo elemento   
print(matriz [-1] [-1]) # último elemento da última linha   
```

#### O fatiamento segue o mesmo padrão das listas

Podemos pegar também partes dos elementos, passando o índice inicial e final, e/ou o step
- Devemos usar os ":" para "acessar"
- Na maioria das vezes devemos passar pelo menos o índice inicial ou final
*Posição final não é retornada*

```python
tupla = ("p", "y", "t", "h", "o", "n",)

print(tupla[:2])
print(tupla[1:3])
print(tupla[0:3:2])
print(tupla[::]) ### usado para copiar ### excessão 
print(tupla[::-1]) ### copia ao contrário
```

#### Iterando tuplas

A forma mais comum é usando "for"

```python
carros = ("gol", "celta", "palio",)

for x in carros:
    print(x)
type(carros)
```

#### Função enumerate

```python
for indice, x in enumerate(carros):
    print(f"{indice}: {x}")
```
