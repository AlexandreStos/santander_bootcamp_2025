### Operadores de associação

#### São operadore utilizados para verificar se um objeto está presente em uma sequência.

Operador "in"

```python
curso = "Curso de Python"
frutas = ["laranja", "uva", "limao"]
saques = [1500,100]
```

```python
"Python" in curso
```

```python
" " in curso ## espaço é caracter
```

```python
"maçã" not in frutas
```

```python
200 in saques 
```

#### lembrar que o operador é case sensitivity
