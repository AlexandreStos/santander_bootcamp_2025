### Operadores lógicos

#### São utilizados conjuntamente com operadores de comparação e retornam um boleano (TRUE or FALSE), utilizados para montar funçoes.

```python
saldo= 1000
saque= 200
limite=100
conta_especial = True
```

### Operador "and"

Ambas as partes da expressão devem ser verdadeiras para retornar "True"

```python
saldo >= saque and saque <= limite 
```

### Operador "or"

umas das partes da expressão (condição) deve ser verdadeira para retornar "True"

```python
saldo >= saque or saque <= limite 
```

```python
### Operador "not"
```

### Esse operador é o oposto do verdadeiro, então dara True

```python
n = 42

if not n:
    print("isso não será executado")
else:
    print("n é considerado verdadeiro.")
```

### Uso de parenteses

```python
(saldo >= saque and saque <= limite) or (conta_especial and saldo >= saque)
```

### Quebrando espressões 

```python
conta_normal_com_saldo_suficiente = saldo >= saque and saque <= limite 
conta_especial_com_saldo_suficiente =  conta_especial and saldo >= saque 
```

```python
exp_1 = conta_normal_com_saldo_suficiente or conta_especial_com_saldo_suficiente
```
