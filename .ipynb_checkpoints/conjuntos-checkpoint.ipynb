### Conjuntos / sets

#### Estrutura de dados set
Um set é uma coleção que não possui objetos repetidos; usamos sets para representar conjuntos matemáticos ou eliminar itens duplicados de um interável. Muitas vezes usado para eliminar a duplicidade

```python
#### set é a função de origem/construtor
```

```python
set([1,2,3,1,3,4,2])
```

```python
set("abacaxi")
```

```python
set(("palio", "gol", "celta", "palio"))## aqui usando tuplas
```

#### Também podemos usar chaves {} para criar sets 

```python
linguagens = {"python", "java", "python"}
```

```python
print(linguagens)
```

#### Conjuntos em python não suportam índice e fatiamento, caso necessário converta para lista list()

```python
numeros = {1,2,3,2}
numeros = list(numeros)
numeros[0]
```

### Podemos interar usando for, como usamos em listas

```python
carros = {"gol", "celta", "palio"}
for x in carros:
    print(x)
```

### Podemos usar o enumerate para descorir o índice de cada objeto dentro do conjunto 

```python
for y, x in enumerate(carros):
    print(f"{y}: {x}")
```

### Métodos para set

#### Todas as propriedades dos conjuntos matemáticos

#### {}.union

```python
conjunto_a = {1,2}
conjunto_b = {3,4}
conjunto_a.union(conjunto_b)
```

#### {}.intersection

```python
conjunto_a = {1,2,3}
conjunto_b = {2,3,4}
conjunto_a.intersection(conjunto_b)
```

#### {}.difference

```python
conjunto_a = {1,2,3}
conjunto_b = {2,3,4}
conjunto_a.difference(conjunto_b)
```

```python
conjunto_b.difference(conjunto_a)
```

#### {}.symmetric_difference

#### Todos os elementos que não estão na intersecção

```python
conjunto_a = {1,2,3}
conjunto_b = {2,3,4}
conjunto_a.symmetric_difference(conjunto_b)
```

### {}.issubset

#### Testa com verdadeiro ou falso se um conjunto pertence ao outro (subset)

```python
conjunto_a = {1,2,3}
conjunto_b = {4,1,2,5,6,3}
```

```python
conjunto_a.issubset(conjunto_b)
```

```python
conjunto_b.issubset(conjunto_a)
```

#### {}.issuperset

Em Python, issuperset() é um método de conjuntos (set). Ele verifica se um conjunto contém todos os elementos de outro conjunto. Se contiver, retorna True. Caso contrário, False. Drama mínimo, resposta binária.

Definição formal (sem poesia)

A.issuperset(B) retorna True se todo elemento de B estiver em A.

Ou seja:

A é um superconjunto de B?

```python
conjunto_a = {1,2,3}
conjunto_b = {4,1,2,5,6,3}
```

```python
conjunto_a.issuperset(conjunto_b)
```

```python
conjunto_b.issuperset(conjunto_a)
```

#### Observações importantes (leia antes de fazer besteira)

Funciona apenas com conjuntos (set ou frozenset)
A ordem dos elementos não importa, porque conjuntos não ligam para isso
Um conjunto sempre é superconjunto de si mesmo

### {}.isdisjoint

#### Confere se todos os conjuntos não estão presentes no outro

```python
conjunto_a = {1,2,3,4,5}
conjunto_b = {6,7,8,9}
conjunto_c = {1,0}
```

```python
conjunto_a.isdisjoint(conjunto_b)
```

```python
conjunto_b.isdisjoint(conjunto_a)
```

```python
conjunto_a.isdisjoint(conjunto_c)
```

### {}.add

#### Definição objetiva
conjunto.add(elemento) adicionará elemento ao conjunto se ele ainda não existir.
Se já existir, nada acontece

```python
a = {1, 2, 3}
a.add(4)
print(a)
```
