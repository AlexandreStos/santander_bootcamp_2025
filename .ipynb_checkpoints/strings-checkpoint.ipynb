#### Maiúsculas, minúsculas e título

```python
curso = "pYtHon"

print(curso.upper()) ## depois de colocar o "." paerte tab e mostrará todos os metodos possíveis para o objeto
```

```python
print(curso.lower())
```

```python
print(curso.title())
```

#### Cortando espaços em branco

#### Dos dois lados

```python
curso = "  python "
print(curso.strip())
```

#### Da esquerda

```python
print(curso.lstrip())
```

### Da direita 

```python
print(curso.rstrip())
```

### Junções e centralização

```python
curso = "Python"
```

```python
print(curso.center(8, "#")) # primeira variavel determina o número de caracteres e a segunda o caractere que será usado externamente. 
```

```python
print("." .join(curso)) # juntando o caracter com a string, primeiro passamos o caracterer e depois passamos sua string, vai passar item a item
```

### Interpolação de variáveis 

#### Old style%

- %s para variável string
- %d para inteiro
- %f para flutuante

### Observação: devemos passar o argumento % e as variáveis, como a seguir: 

### Ex.: 

```python
nome = "Guilherme"
idade = 28
linguagem = "Python"
profissao = "Programador"
```

```python
print("Ola, me chamo %s. Eu tenho %d anos de idade, trabalho como %s e estou matriculado no curso de %s." % (nome, idade, profissao, linguagem))
```

#### Método format

#### Observação: devemos passar o argumento .format() e as variáveis, no texto devemos abrir chaves, além disso, podemos passar o indice (como referências) das variáveis, como a seguir: 

```python
print("Olá, me chamo {0}. Eu tenho {1} anos de idade, trabalho como {2} e estou matriculado no curso de {3}." .format (nome, idade, profissao, linguagem))
```

#### Podemos passar também como argumento o nome da váriavel

```python
print("Olá, me chamo {nome}. Eu tenho {idade} anos de idade, trabalho como {profissao} e estou matriculado no curso de {linguagem}." 
      .format (nome = "alexandre", idade = 40, profissao = "cientista", linguagem = "R"))
```

### f-string (mais recente)

```python
nome = "Guilherme"
idade = 28
linguagem = "Python"
profissao = "Programador"
```

```python
print(f"Olá, me chamo {nome}. Eu tenho {idade} anos de idade, trabalho como {profissao} e estou matriculado no curso de {linguagem}.") 
```

#### Formatação de float de string com f-string

```python
PI = 3.14159
```

```python
print(f"Valor de PI: {PI:.2f}") ## duas casas decimais 
```

```python
print(f"Valor de PI: {PI:10.3f}") ## três casas (f=float) decimais; o 10 é dez espaços (totais)
```

### Trabalhando com dicionários, format e strings

```python
dados = {"nome": "Gui", "idade": 28}

print("Nome: {nome}; Idade: {idade}".format(**dados)) ### passamos o dicionário
```

### Fatiamento

#### Técnica utilizada para retornar substrings, informando o inicio, fim, e passo [start:stop:step]

### Ex. utilizando índice:

```python
nome = "Alexandre Moura dos Santos"
```

```python
nome[0]
```

```python
nome[:9] 
### Se não colocar o primeiro argumento ele vai considerar iniciar no indice 0, primeiro argumento é o start
```

```python
nome[10:]Se não colocar o segundo argumento ele irá até o fim
```

```python
nome[10:15]
```

```python
nome[10:15:2]## pega de dois em dois
```

```python
nome[:] ## Pega tudo
```

```python
nome[::-1] ## espelhamento
```

### Strings de multiplas linhas/ strings triplas

#### Sao informadas usando 3 aspas duplas ou triplas

```python
mensagem = f"""
##########################################################
  Olá meu nome é {nome}, eu estou estudando python.
Tenho 40 anos
##########################################################
"""
print(mensagem)   ### mantem a formatação de espaçamento
```
