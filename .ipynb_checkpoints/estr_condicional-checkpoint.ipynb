### Identação e os blocos 

A identaçao serve para organizar e mostrar onde o bloco de código começa ou termina, melhorando a visualização do código, parecendo com paragráfos.

**O PEP 8 define regras para a indentação e o espaçamento em Python, a fim de tornar o código mais legível. Aqui estão algumas diretrizes importantes:**
- Use 4 espaços para cada nível de indentação (não use tabulações).

- Não misture espaços e tabulações para indentação.

- Evite linhas em branco no início e no final do bloco de código.

- Separe as funções e classes com duas linhas em branco.

- Use espaços em branco ao redor de operadores e após vírgulas.

Para cada bloco devemos dar mais 4 espaços, Exemplo: 

```python
def sacar (valor):
    saldo = 500

   
    if saldo >= valor:
        print("valor sacado")
    else:
        print("sem saldo")
```

```python
sacar(600)
```

### Estruturas condicionais 

Permite o desvio de fluxo quando alguma condição é atendida

#### Estrutura condicional simple "if"

É utilizado quando o objetivo é criar um único desvio, caso a expressão lógica seja atendida, rodará o bloco de código já descrito. Por exemplo: 

```python
saldo =2000
saque = float(input("Qual o valor de saque em reais?"))
if saque <= 2000:
    print(f"O valor de R$ {saque} foi liberado")
if saque > 2000:
    print(f"O valor de R$ {saque} não foi liberado, limite excedido!")
    
```

#### Estrutura condicional if/else

Para dois desvios, o bloco será executado em ordem conforme atendida a condição, condição deve ser diretamente oposta.

```python
saldo =2000
saque = float(input("Qual o valor de saque em reais?"))
if saque <= 2000:
    print(f"O valor de R$ {saque} foi liberado")
else:
    print(f"O valor de R$ {saque} não foi liberado, limite excedido!")
```

#### Estrutura condicional elif

 Usado quando temos necessidade de mais de duas condições, os elif são usados no meio entre if e else e podem ser infinitos em quantidade

```python
import sys
opcao = int(input("Informe uma opção: [1] Sacar \n[2] Extrato: "))
if opcao == 1:
    saldo =2000
    saque = float(input("Qual o valor de saque em reais?"))
    if saque <= 2000:
        print(f"O valor de R$ {saque} foi liberado")
    else:
        print(f"O valor de R$ {saque} não foi liberado, limite excedido!")
elif opcao == 2:
    print("Exibindo o extrato...")
else: 
    sys.exit("Opção inválida")
                  
```

#### If aninhado

#####  Posso criar estruturas if/elif/else dentro de outras estruturas if, como descrito no código.

```python
import sys
opcao = int(input("Informe uma opção: [1] Sacar \n[2] Extrato: "))
if opcao == 1:
    saldo =2000
    saque = float(input("Qual o valor de saque em reais?"))
#### Essa parte ### 
    if saque <= 2000:
        print(f"O valor de R$ {saque} foi liberado")
    else:
        print(f"O valor de R$ {saque} não foi liberado, limite excedido!")
#####
elif opcao == 2:
    print("Exibindo o extrato...")
else: 
    sys.exit("Opção inválida")
                  
```

#### if ternário

Permite escrever uma condição em uma única linha, temos a ordem: se if for atendido resposta do inicio, condição, e resposta se não for atendido.
Ex:
valor_se_verdadeiro if condição else valor_se_falso

```python
#### If aninhado
saldo =2000
saque = float(input("Qual o valor de saque em reais?"))
status = "Sucesso" if saldo >= saque else "Falha"
print(f"{status} ao realizar o saque")
```
