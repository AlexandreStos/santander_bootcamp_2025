### Estruturas de Repetição

##### Estruturas for e while
     São utilizadas para repetir um treçho de código um determinado número de vezes, podendo determinar este número previamente ou por lógica

#### Exemplo 1 (estrutura da repetição): 

### Estrutura for 

##### Utilizado quando sabemos o número exato de repetições ou temos o objeto já determinado que será interado

```python
texto = input("Informe um texto: ")
VOGAIS = "AEIOU" ## constante com letra maíuscula
for letra in texto:
    if letra.upper() in VOGAIS:    ### .upper converte para maiusculo 
        print(letra, end="")
print() # adiciona uma quebra de linha
```

### Estrutura for else
     ##### Pouco usado na prática

```python
texto = input("Informe um texto: ")
VOGAIS = "AEIOU" ## constante com letra maíuscula
for letra in texto:
    if letra.upper() in VOGAIS:    ### .upper converte para maiusculo 
        print(letra, end="")
else: 
    print() # adiciona uma quebra de linha
    print("Executa ao final do laço")

```

### Range

É uma função interna do python que determina um ínicio e um final para interação. Recebendo 3 argumentos stop, start e step. 

#### *Importante: quando usamos a função o primeiro número interável é incluido e o último descrito na função não entra na interação*


```python
list(range(0,10))  ## até o nono número, podemos usar ou não o argumento start quando começamos em zero
```

```python
list(range(10))  ## até o nono número, sem o start
```

### Uso do range em repetições

```python
for numero in range(0,11):
    print(numero, end=", ")## end coloca como deverão ser separados
```

```python
# exibindo a tabuada do cinco
for numero in range(0,51,5): ### o terceiro argumento é step, ou seja, o passo da repetição
    print(numero, end="; ")## end inseri como deverão ser separados
```

### Interador while

#####  Usado para interar até que alguma condição aconteça, sem saber a quantidade exata de interações


```python
opcao = -1
while opcao != 0:
    opcao = int(input("[1] Sacar \n[2] Extrato \n[0] Sair \n: "))
    if opcao == 1:
        print("Sacando...")
    elif opcao == 2:
        print("Exibindo extrato...")
    elif opcao >= 3:
        print("Opção incorreta, digite novamente")
else:
    print("obrigado por utilizar")
```

### Break

#### para obrigar uma parada 

```python
opcao = -1
while True: ### laço infinito
    numero = int(input("iNFORME UM NÚMERO: "))
    if numero == 10:
        break
    print(numero)
```

```python
import random

random.seed(42)              # fixa a seed
SENHA = random.randint(1, 10)

numero = int(input("Digite seu palpite: "))

if numero == SENHA:
    print("Acertou a senha!")
else:
    print("Senha incorreta. A senha era", SENHA)
```

### Continue

```python
#### Usado para pular um número no laço, ou evento..
```

```python
for numero in range(100):
    if numero ==12:
        continue
    print(numero, end=" ")
```

```python
#### printando os números impares 
```

```python
for numero in range(100):
    if numero % 2 == 0:
        continue
    print(numero, end=" ")
```
