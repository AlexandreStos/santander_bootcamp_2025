**OU**

##### habilitamos o módulo math e usamos a função específica 

```python
import math
numero = 16
raiz = math.sqrt(numero)
print(f"A raiz quadrada de {numero} é {raiz}")
```

#### Raiz cúbica

```python
numero = 27
raiz_cubica = numero ** (1/3)  # ou pow(numero, 1/3)
print(f"A raiz cúbica de {numero} é {raiz_cubica}")
```

#### Raiz para numeros negativos ou complexos

```python
import cmath
numero = -25
raiz = cmath.sqrt(numero)
print(f"A raiz quadrada de {numero} é {raiz}")
```
